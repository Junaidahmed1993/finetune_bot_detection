```python
# # now, Lets concatc all the files

# # List of columns to keep
# columns_to_keep = ['text_data', 'Label']  # Add other columns you want to keep

# # Concatenate all DataFrames vertically
# merged_df = pd.concat([df1[columns_to_keep], df2[columns_to_keep], df3[columns_to_keep], df4[columns_to_keep], df5[columns_to_keep]], ignore_index=True)

# # Check the merged DataFrame
# print(merged_df.head())
import pandas as pd
import numpy as np

# Assuming df1, df2, df3, df4, df5 are your DataFrames loaded from CSV files

# List of columns to keep
columns_to_keep = ['text_data', 'Label']  # Add other columns you want to keep

# Filter data into human and bot categories
human_data = pd.concat([df[df['Label'] == 'human'] for df in [df1, df2, df3, df4, df5]], ignore_index=True)
bot_data = pd.concat([df[df['Label'] == 'bot'] for df in [df1, df2, df3, df4, df5]], ignore_index=True)

# Determine the number of samples to select from each category
num_samples_per_category = 25000

# Randomly select equal numbers of samples from each category
human_indices = np.random.choice(human_data.index, size=num_samples_per_category, replace=False)
bot_indices = np.random.choice(bot_data.index, size=num_samples_per_category, replace=False)

# Subset the data using the randomly selected indices
merged_df = pd.concat([human_data.loc[human_indices, columns_to_keep], bot_data.loc[bot_indices, columns_to_keep]], ignore_index=True)

# Shuffle the merged DataFrame
merged_df = merged_df.sample(frac=1).reset_index(drop=True)

# Check the merged DataFrame
print(merged_df.head())


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Label' is your label column in the final_df DataFrame

# Plotting the distribution of labels
sns.countplot(x='Label', data=merged_df)
plt.title('Distribution of Labels')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
# Define the path where you want to save the merged CSV file
output_file_path = "/content/drive/My Drive/Ok Data/merged_data_equal.csv"

# Save the merged DataFrame to a CSV file
merged_df.to_csv(output_file_path, index=False)

print("Merged data saved successfully.")
```
